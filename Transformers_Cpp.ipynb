```python
%%writefile "/content/drive/MyDrive/Transformer C++/src/Tensor.hpp"
#include <vector>
#include <iostream>
#include <cassert>

class Tensor {
private:
    std::vector<float> data;
    std::vector<int> shape;

public:
    Tensor(std::vector<int> shape) : shape(shape) {
        int total = 1;
        for (int s : shape) total *= s;
        data.resize(total, 0.0f);
    }

    int size() const {
        return data.size();
    }

    std::vector<int> get_shape() const {
        return shape;
    }

    float& operator[](int idx) {
        return data[idx];
    }

    const float& operator[](int idx) const {
        return data[idx];
    }

    std::vector<float>& raw() {
        return data;
    }
    float& at(int i, int j) {
      int cols = shape[1];
      return data[i * cols + j];
    }
    Tensor add(const Tensor& other) {
      assert(data.size() == other.data.size());

      Tensor result(shape);

      for (int i = 0; i < data.size(); i++) {
          result[i] = data[i] + other[i];
      }

      return result;
    }
    Tensor matmul(const Tensor& other) {
    int m = shape[0];
    int n = shape[1];
    int p = other.shape[1];

    Tensor result({m, p});

    for (int i = 0; i < m; i++) {
        for (int j = 0; j < p; j++) {
            float sum = 0.0;

            for (int k = 0; k < n; k++) {
                sum += at(i, k) * other.at(k, j);
            }

            result.at(i, j) = sum;
        }
    }

      return result;
    }
};
```

```python
%%writefile "/content/drive/MyDrive/Transformer C++/include/Linear.cpp"
#include <vector>
#include <random>
#include <cmath>
#include <cassert>
#include"Linear.hpp"
class Linear {
private:
    int in_dim, out_dim;
    std::vector<float> weight; // (out_dim x in_dim)
    std::vector<float> bias;

public:
    Linear(int in_dim, int out_dim)
        : in_dim(in_dim), out_dim(out_dim) {

        weight.resize(out_dim * in_dim);
        bias.resize(out_dim);

        initialize();
    }

    void initialize() {
        std::mt19937 gen(42);
        std::normal_distribution<float> dist(0.0, 0.02);

        for (auto &w : weight) w = dist(gen);
        for (auto &b : bias) b = 0.0f;
    }

    float& W(int i, int j) {
        return weight[i * in_dim + j];
    }

    // X: (seq_len x in_dim)
    Tensor forward(const Tensor& X) {
        int seq_len = X.get_shape()[0];

        Tensor output({seq_len, out_dim});

        for (int i = 0; i < seq_len; i++) {
            for (int j = 0; j < out_dim; j++) {

                float sum = bias[j];

                for (int k = 0; k < in_dim; k++) {
                    sum += X.at(i, k) * W(j, k);
                }

                output.at(i, j) = sum;
            }
        }

        return output;
    }
};
```

```python
%%writefile "/content/drive/MyDrive/Transformer C++/include/SelfAttention.cpp"
class SelfAttentionProjections {
public:
    Linear Wq, Wk, Wv;

    SelfAttentionProjections(int d_model)
        : Wq(d_model, d_model),
          Wk(d_model, d_model),
          Wv(d_model, d_model) {}

    void forward(const Tensor& X, Tensor& Q, Tensor& K, Tensor& V) {
        Q = Wq.forward(X);
        K = Wk.forward(X);
        V = Wv.forward(X);
    }
};
```

```python
%%writefile "/content/drive/MyDrive/Transformer C++/include/Transpose.cpp"
Tensor transpose(const Tensor& X) {
    int rows = X.get_shape()[0];
    int cols = X.get_shape()[1];

    Tensor result({cols, rows});

    for (int i = 0; i < rows; i++) {
        for (int j = 0; j < cols; j++) {
            result.at(j, i) = X.at(i, j);
        }
    }

    return result;
}
```

```python
%%writefile "/content/drive/MyDrive/Transformer C++/include/Softmax.cpp"
Tensor softmax(const Tensor& X) {
    int rows = X.get_shape()[0];
    int cols = X.get_shape()[1];

    Tensor result({rows, cols});

    for (int i = 0; i < rows; i++) {
        float max_val = -1e9;

        for (int j = 0; j < cols; j++) {
            max_val = std::max(max_val, X.at(i, j));
        }

        float sum = 0.0;

        for (int j = 0; j < cols; j++) {
            float val = std::exp(X.at(i, j) - max_val);
            result.at(i, j) = val;
            sum += val;
        }

        for (int j = 0; j < cols; j++) {
            result.at(i, j) /= sum;
        }
    }

    return result;
}
```

```python
%%writefile "/content/drive/MyDrive/Transformer C++/include/Attention.cpp"
Tensor attention(const Tensor& Q, const Tensor& K, const Tensor& V) {
    int d_k = Q.get_shape()[1];
    // K^t
    Tensor K_T = transpose(K);
    // QK^T
    Tensor scores = Q.matmul(K_T);

    float scale = 1.0 / std::sqrt(d_k);
    // QK^T/√D
    for (int i = 0; i < scores.size(); i++) {
        scores[i] *= scale;
    }
    // softmax(QK^T/√D)
    Tensor weights = softmax(scores);
    // softmax(QK^T/√D)*V
    Tensor output = weights.matmul(V);

    return output;
}
```
